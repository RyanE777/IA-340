Group and Visualize the Data: View Below

Markdown: 
Plot 1: For the first plot, I decided to create a bar chart to see the Electric Range for each Car brand Model. We can see from this chart that a few Models are Better than others.

Plot 2: For my second plot, I decided to create a line chart to see the Electric Range for each model of Vehicle. We can see from this chart that the Model 3 outplays the other Models of Electric Vehicles by far.

```python

sum_range_by_model = df.groupby('Model')['Electric Range'].sum()
sum_range_by_model.plot.bar(ylabel = 'Electric Range')
```

```python
sum_range_by_make = df.groupby('Model')['Electric Range'].sum()
sum_range_by_make.plot(ylabel = 'Electric Range')
```
